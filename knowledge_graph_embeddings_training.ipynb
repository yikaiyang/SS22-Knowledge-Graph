```python
from pykeen.triples import TriplesFactory 
from pykeen.pipeline import pipeline
```

```python
tf = TriplesFactory.from_path_binary('/Users/yikaiyang/Projects/SS22-Knowledge-Graph/2_knowledge-graph/triples')
training, testing, validation = tf.split([.8, .1, .1])
```

```python
result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model='TransE',
    training_kwargs=dict(
        num_epochs=2000,
        checkpoint_name='kge_transe.pt',
        checkpoint_directory='checkpoint/'
    ),
)
```

```python
result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model='TransE',
    training_kwargs=dict(
        num_epochs=2000,
        checkpoint_name='kge_transe.pt',
        checkpoint_directory='checkpoint/'
    ),
)
```

```python
result.plot(figsize=(40, 16))
```
